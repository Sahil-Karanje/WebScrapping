# Internshala Scrapping

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://internshala.com/internships/data-science-internship/"
data = requests.get(url).text
```

```python
dictData = {"Title":[],"Company Name":[],"Location":[],"Start Date":[],"Duration":[],"Stipend":[]}
```

```python
soup = BeautifulSoup(data,"html.parser")
for i in soup.select("h3.heading_4_5.profile"):
    dictData["Title"].append(i.get_text().strip())
```

```python
for i in soup.select("div.company_and_premium"):
    dictData["Company Name"].append(i.get_text().strip())
```

```python
for i in soup.select("div#location_names"):
    dictData["Location"].append(i.get_text().strip())
```

```python
for i in soup.select("span.start_immediately_desktop"):
    dictData["Start Date"].append(i.get_text().strip())
```

```python
rows = soup.find_all('div', class_='other_detail_item_row')

for row in rows:
    item_bodies = row.find_all('div', class_='item_body')
    for item_body in item_bodies:
        if 'Months' in item_body.get_text() or 'Weeks' in item_body.get_text():
            duration_text = item_body.get_text(strip=True)
            dictData["Duration"].append(i.get_text().strip())
```

```python
for i in soup.select("span.stipend"):
    dictData["Stipend"].append(i.get_text().strip())
```

```python
print(len(dictData["Title"]))
print(len(dictData["Company Name"]))
print(len(dictData["Location"]))
print(len(dictData["Start Date"]))
print(len(dictData["Duration"]))
print(len(dictData["Stipend"]))
```

```python
df = pd.DataFrame(dictData)
df.to_csv("InternShala.csv",index=False,encoding='utf-8-sig')
```
